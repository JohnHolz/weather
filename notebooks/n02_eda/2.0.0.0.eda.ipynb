## EDA - Exploratory Data Analysis
Dataset 0 is the Univariated dataset from our data.  
The only variable in the dataset 0 is 'maximum temperature' in station A612 - Vitoria, capital of Espirito Santo.

### Importing Libs

```python
## basic libs
import pandas as pd
import numpy as np

## ploting libs
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

figsize = (20,8)
```

### Reading data

```python
df = pd.read_csv('../../data/04_datasets/dataset_0.csv')
dataset = df.iloc[:,1:].iloc[:,:-6]
df = df.iloc[:,:2]
df.date_time = pd.to_datetime(df.date_time)
df['month'] = df.date_time.dt.month_name()
df['year'] = df.date_time.dt.year
df['hour'] = df.date_time.dt.hour
```

```python
_ = df[df["date_time"] >= '2015-01-01'].resample('3d', on='date_time').mean().iloc[:,0].plot(figsize= figsize)
```

```python
plt.figure(figsize=figsize)
_ = sns.boxplot(data=df,y = 'A612 - temperatura maxima na hora ant. (aut) (°c)', x='month')
```

```python
plt.figure(figsize=figsize)
_ = sns.boxplot(data=df,y = 'A612 - temperatura maxima na hora ant. (aut) (°c)', x='year')
```

```python
plt.figure(figsize=figsize)
_ = sns.histplot(df['A612 - temperatura maxima na hora ant. (aut) (°c)'])
```
