```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
n = 500
e_mean = 0
e_sd   = 2
```

```python
e = np.random.normal(e_mean, e_sd,size = n)
```

```python
x = [1,2,3,4,5]
```

```python
x[1:]
```

```python
def ar1(e, param1 = 0.5):
    ret = [e[0]]

    for i in range(1,len(e)):
        ret.append(ret[i-1]*param1 + e[i])
        
    return ret
```

```python
ar = ar1(e,1.001)#0.99)
```

```python
_ = sns.lineplot(y = ar,
                    x = range(len(e)),markers=True)
```
